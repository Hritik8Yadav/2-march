```python
#Ans2: Scatter plot is a plot which plots the data on x and y axis in the form of points.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.figure(figsize=(3,2))
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Chart")
```

```python
#Ans3: subplot is a function of matplotlib used for multiple plot in same graph.
#For line 1: 
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(1,2,1)
plt.plot(x,y)
#For line 2: 
y1 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y1)
#For line 3: 
y2 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y2)
#For line 4: 
y3 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y3)
```

```python
#Ans4: bar plot is a type of plot used to plot the categorical data on a graph.
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```
